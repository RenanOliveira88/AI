```python
import cv2
import matplotlib.pyplot as plt
```

```python
imagem = cv2.imread("a.jpg")
plt.imshow(imagem)
```

```python
imagem = cv2.cvtColor(imagem, cv2.COLOR_RGB2BGR)
plt.imshow(imagem)
```

```python
imgRavel = imagem.ravel()
plt.hist(imgRavel)
```

```python
plt.clf()
imagemPb = cv2.imread("a.jpg", 0)
plt.imshow(imagemPb)
```

```python
imgRavel = imagemPb.ravel()
plt.hist(imgRavel)
```

```python
plt.clf()
equ = cv2.equalizeHist(imagemPb)
plt.imshow(equ)
```

```python
equRavel = equ.ravel()
plt.hist(equRavel)
```
