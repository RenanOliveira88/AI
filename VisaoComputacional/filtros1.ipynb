```python
import cv2 as cv
from matplotlib import pyplot as plt
import numpy as np
```

```python
img = cv.imread("a.jpg")
plt.imshow(img)
```

```python
kernel = np.ones((5,5),np.float32)/25
```

```python
dst = cv.filter2D(img,-1,kernel)
plt.imshow(dst)
```

```python
plt.subplot(121),plt.imshow(img), plt.title("original")
plt.subplot(122),plt.imshow(dst), plt.title("borrado")
```

```python
plt.show()
plt.clf()
```

```python
blur = cv.blur(img,(5,5))
plt.imshow(blur)
```

# Erosão e dilatação

```python
struct = np.ones((5,5), np.uint8)
```

```python
img_erosao = cv.erode(img, struct)
plt.imshow(img_erosao)
```

```python
img_dilatacao = cv.dilate(img, struct)
plt.imshow(img_dilatacao)
```
